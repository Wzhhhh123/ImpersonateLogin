```python
from selenium import webdriver
from time import sleep

driver = webdriver.Firefox()
driver.get("https://movie.douban.com/tag/#/?sort=U&range=0,10&tags=%E7%94%B5%E5%BD%B1,%E4%B8%AD%E5%9B%BD%E5%A4%A7%E9%99%86")
```

```python
movie_synopsis=[]
movie_name=[]
movie_jianjie=[]
```

```python
def yue():
    driver.find_element_by_xpath("/html/body/div[3]/div[1]/div/div[1]/div/div/div[1]/a").click()
    sleep(3)
   
```

```python
k=0  #填写上次中断的条数

try:
    tiaoshu = int(input("想要爬取的条数："))
except:
    print("请输入整数数字")
    tiaoshu = int(input("想要爬取的条数："))

for i in range(tiaoshu):
    k=k+1
    print("第",k,"条")
    try:
        driver.find_element_by_xpath("/html/body/div[3]/div[1]/div/div[1]/div/div/div[1]/div[3]/a[{}]/p/span[1]".format(k)).click()
    except:
        try:
            yue()
            driver.find_element_by_xpath("/html/body/div[3]/div[1]/div/div[1]/div/div/div[1]/div[3]/a[{}]/p/span[1]".format(k)).click()
        except:
            for i in range(k//20):
                yue()
            driver.find_element_by_xpath("/html/body/div[3]/div[1]/div/div[1]/div/div/div[1]/div[3]/a[{}]/p/span[1]".format(k)).click()
    all_h = driver.window_handles
    driver.switch_to.window(all_h[1])
    sleep(2)
    try:
        name1=driver.find_element_by_xpath("/html/body/div[3]/div[1]/h1/span[1]").text
        print("电影名：",name1)
        padao=driver.find_element_by_id("info").text
        movie_synopsis.append(padao)
        movie_name.append(name1)
    except:
        try:
            sleep(10)
            name1=driver.find_element_by_xpath("/html/body/div[3]/div[1]/h1/span[1]").text
            print("电影名：",name1)
            padao=driver.find_element_by_id("info").text
            movie_synopsis.append(padao)
            movie_name.append(name1)
        except:
            try:
                sleep(30)
                name1=driver.find_element_by_xpath("/html/body/div[3]/div[1]/h1/span[1]").text
                print("电影名：",name1)
                padao=driver.find_element_by_id("info").text
                movie_synopsis.append(padao)
                movie_name.append(name1)
            except:
                try:
                    sleep(90)
                    name1=driver.find_element_by_xpath("/html/body/div[3]/div[1]/h1/span[1]").text
                    print("电影名：",name1)
                    padao=driver.find_element_by_id("info").text
                    movie_synopsis.append(padao)
                    movie_name.append(name1)
                except:
                    print("输入任意字符回车，继续爬取,快去换代理IP！")
                    sdwadw = input("输入：")
                    sleep(5)
                    name1=driver.find_element_by_xpath("/html/body/div[3]/div[1]/h1/span[1]").text
                    print("电影名：",name1)
                    padao=driver.find_element_by_id("info").text
                    movie_synopsis.append(padao)
                    movie_name.append(name1)
                
    try:
        jianjie=driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[2]/div[1]/div[3]/div/span").text
        if (jianjie !=' ' or jianjie !=''or jianjie !='  '):
            movie_jianjie.append(jianjie)
        else:
            jianjie=driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[3]/div[1]/div[3]/div/span[1]").text
            if (jianjie !=' ' or jianjie !=''or jianjie !='  '):
                movie_jianjie.append(jianjie)
    except:
        try:
            jianjie=driver.find_element_by_xpath("/html/body/div[3]/div[1]/div[3]/div[1]/div[3]/div/span[1]").text
            movie_jianjie.append(jianjie)
        except:
            jianjie='null'
            movie_jianjie.append(jianjie)
    print(padao)
    print("简介：",jianjie)
    driver.close()
    driver.switch_to.window(all_h[0])
```
